```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función 3D
def function_3d(x, y, z):
    return x**2 + y**2 + z**2

# Calcular el gradiente de la función
def gradient_3d(x, y, z):
    df_dx = 2 * x
    df_dy = 2 * y
    df_dz = 2 * z
    return np.array([df_dx, df_dy, df_dz])

# Algoritmo de descenso del gradiente
def gradient_descent_3d(learning_rate, iterations):
    # Inicializar valores
    x = 5.0
    y = 5.0
    z = 5.0
    history = []

    # Iteraciones del descenso del gradiente
    for _ in range(iterations):
        value = function_3d(x, y, z)
        history.append((x, y, z, value))

        # Calcular el gradiente
        gradient = gradient_3d(x, y, z)

        # Actualizar los parámetros
        x -= learning_rate * gradient[0]
        y -= learning_rate * gradient[1]
        z -= learning_rate * gradient[2]

    return history

# Configuración del descenso del gradiente
learning_rate = 0.1
iterations = 100

# Ejecutar el descenso del gradiente
history = gradient_descent_3d(learning_rate, iterations)

# Visualización de la función y el descenso del gradiente
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear valores para x, y, z
x_vals = np.linspace(-6, 6, 100)
y_vals = np.linspace(-6, 6, 100)
z_vals = np.linspace(-6, 6, 100)

# Crear mallas 3D
X, Y, Z = np.meshgrid(x_vals, y_vals, z_vals)

# Calcular los valores de la función en la malla
F = function_3d(X, Y, Z)

# Plot de la superficie
#ax.plot_surface(X, Y, F, alpha=0.5, cmap='viridis', edgecolors='k', label='Función 3D')

# Extraer valores de la historia del descenso del gradiente
x_history, y_history, z_history, _ = zip(*history)
ax.plot(x_history, y_history, z_history, color='red', marker='o', label='Descenso del Gradiente')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Función 3D y Descenso del Gradiente')

plt.legend()
plt.show()


```

Teóricamente, el descenso del gradiente y la retropropagación (backpropagation) están estrechamente relacionados en el contexto de las redes neuronales. Aquí hay una explicación general de cómo se relacionan:

1. **Descenso del Gradiente:**
   - El descenso del gradiente es un algoritmo de optimización utilizado para minimizar una función de costo. En el contexto de las redes neuronales, esta función de costo mide qué tan bien la red está realizando la tarea deseada en comparación con los resultados reales.
   - El objetivo es ajustar los pesos y sesgos de la red para minimizar la función de costo. Esto se logra iterativamente actualizando los parámetros en la dirección opuesta al gradiente de la función de costo con respecto a esos parámetros.

2. **Retropropagación (Backpropagation):**
   - La retropropagación es un algoritmo específico utilizado para calcular el gradiente de la función de costo con respecto a los pesos de la red. Este cálculo se realiza eficientemente utilizando la regla de la cadena.
   - La retropropagación se llama así porque el gradiente se calcula retrocediendo a través de la red neuronal, de la capa de salida hacia la capa de entrada. Durante este proceso, se calculan los gradientes parciales para cada peso y sesgo.

3. **Conexión Teórica:**
   - En el entrenamiento de una red neuronal, se utiliza el descenso del gradiente para actualizar los pesos y sesgos de la red con el objetivo de minimizar la función de costo.
   - La retropropagación se utiliza para calcular el gradiente de la función de costo con respecto a los pesos y sesgos de la red, proporcionando así la información necesaria para la actualización de los parámetros en el descenso del gradiente.

En resumen, la retropropagación es un componente esencial en el entrenamiento de redes neuronales, y su papel fundamental es proporcionar los gradientes necesarios para que el descenso del gradiente pueda ajustar los parámetros de la red de manera eficiente. La retropropagación y el descenso del gradiente trabajan juntos para mejorar el rendimiento de la red neuronal durante el entrenamiento.

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Definir una red neuronal simple
model = models.Sequential([
    layers.Dense(2, input_shape=(2,), activation='relu'),  # Capa oculta con 2 neuronas y activación ReLU
    layers.Dense(1, activation='sigmoid')  # Capa de salida con 1 neurona y activación sigmoide para problemas de clasificación binaria
])

# Definir una función de costo (por ejemplo, para un problema de clasificación binaria)
loss_fn = tf.keras.losses.BinaryCrossentropy()

# Datos de ejemplo
X_train = tf.constant([[0.5, 0.6]])
y_true = tf.constant([[1.0]])

# Paso de entrenamiento
with tf.GradientTape() as tape:
    # Propagación hacia adelante
    y_pred = model(X_train)

    # Calcular la pérdida
    loss = loss_fn(y_true, y_pred)

# Obtener gradientes usando la retropropagación
gradients = tape.gradient(loss, model.trainable_variables)

# Aplicar descenso del gradiente para actualizar los pesos
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
optimizer.apply_gradients(zip(gradients, model.trainable_variables))

# Imprimir los pesos antes y después de la actualización
print("Pesos antes de la actualización:")
print(model.get_weights())

# Imprimir la pérdida
print("Pérdida:", loss.numpy())

# Entrenamiento completo
for _ in range(100):
    with tf.GradientTape() as tape:
        y_pred = model(X_train)
        loss = loss_fn(y_true, y_pred)

    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

# Imprimir los pesos después del entrenamiento
print("\nPesos después del entrenamiento:")
print(model.get_weights())

# Imprimir la pérdida después del entrenamiento
print("Pérdida después del entrenamiento:", loss.numpy())
```

```python
import numpy as np

# Funciones de activación y sus derivadas
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Datos de entrada y salida
X = np.array([[0.5, 0.6]])
y_true = np.array([[1]])

# Parámetros de la red neuronal (inicializados aleatoriamente)
np.random.seed(42)
input_size = 2
hidden_size = 2
output_size = 1

weights_input_hidden = np.random.rand(input_size, hidden_size)
weights_hidden_output = np.random.rand(hidden_size, output_size)

# Hiperparámetros
learning_rate = 0.01
epochs = 1

# Entrenamiento de la red neuronal
for epoch in range(epochs):
    # Propagación hacia adelante
    hidden_input = np.dot(X, weights_input_hidden)
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, weights_hidden_output)
    y_pred = sigmoid(final_input)

    # Calcular la pérdida
    loss = 0.5 * np.sum((y_pred - y_true)**2)

    # Retropropagación
    error = y_pred - y_true
    delta_output = error * sigmoid_derivative(y_pred)

    error_hidden = delta_output.dot(weights_hidden_output.T)
    delta_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Actualizar pesos con descenso del gradiente
    weights_hidden_output -= learning_rate * hidden_output.T.dot(delta_output)
    weights_input_hidden -= learning_rate * X.T.dot(delta_hidden)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Imprimir los pesos después del entrenamiento
print("\nPesos después del entrenamiento:")
print("Pesos input-hidden:")
print(weights_input_hidden)
print("Pesos hidden-output:")
print(weights_hidden_output)
```

La elección de utilizar `x * (1 - x)` en lugar de `sigmoid(x) * (1 - sigmoid(x))` en el cálculo de la derivada de la función de activación sigmoide durante la retropropagación es una convención común. Ambas formas son equivalentes y pueden utilizarse, pero `x * (1 - x)` es una versión algebraicamente más eficiente de la derivada de la sigmoide.

La derivada de la función sigmoide es:

\[ \sigma'(x) = \sigma(x) \cdot (1 - \sigma(x)) \]

donde \(\sigma(x)\) es la función sigmoide. Si representamos \(\sigma(x)\) como `y`, la derivada se convierte en:

\[ \sigma'(x) = y \cdot (1 - y) \]

Cuando estamos en la capa de salida de una red neuronal y ya hemos calculado la salida de la función sigmoide (que llamamos `y`), podemos simplemente usar `y * (1 - y)` en lugar de llamar nuevamente a la función sigmoide para mejorar la eficiencia computacional.

Ambas formas son correctas y darán los mismos resultados, pero `x * (1 - x)` es una forma más eficiente de calcular la derivada si ya hemos calculado la salida de la función sigmoide en la capa de salida.

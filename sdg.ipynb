```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin 3D
def function_3d(x, y, z):
    return x**2 + y**2 + z**2

# Calcular el gradiente de la funcin
def gradient_3d(x, y, z):
    df_dx = 2 * x
    df_dy = 2 * y
    df_dz = 2 * z
    return np.array([df_dx, df_dy, df_dz])

# Algoritmo de descenso del gradiente
def gradient_descent_3d(learning_rate, iterations):
    # Inicializar valores
    x = 5.0
    y = 5.0
    z = 5.0
    history = []

    # Iteraciones del descenso del gradiente
    for _ in range(iterations):
        value = function_3d(x, y, z)
        history.append((x, y, z, value))

        # Calcular el gradiente
        gradient = gradient_3d(x, y, z)

        # Actualizar los parmetros
        x -= learning_rate * gradient[0]
        y -= learning_rate * gradient[1]
        z -= learning_rate * gradient[2]

    return history

# Configuracin del descenso del gradiente
learning_rate = 0.1
iterations = 100

# Ejecutar el descenso del gradiente
history = gradient_descent_3d(learning_rate, iterations)

# Visualizacin de la funcin y el descenso del gradiente
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear valores para x, y, z
x_vals = np.linspace(-6, 6, 100)
y_vals = np.linspace(-6, 6, 100)
z_vals = np.linspace(-6, 6, 100)

# Crear mallas 3D
X, Y, Z = np.meshgrid(x_vals, y_vals, z_vals)

# Calcular los valores de la funcin en la malla
F = function_3d(X, Y, Z)

# Plot de la superficie
#ax.plot_surface(X, Y, F, alpha=0.5, cmap='viridis', edgecolors='k', label='Funcin 3D')

# Extraer valores de la historia del descenso del gradiente
x_history, y_history, z_history, _ = zip(*history)
ax.plot(x_history, y_history, z_history, color='red', marker='o', label='Descenso del Gradiente')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Funcin 3D y Descenso del Gradiente')

plt.legend()
plt.show()


```

Tericamente, el descenso del gradiente y la retropropagacin (backpropagation) estn estrechamente relacionados en el contexto de las redes neuronales. Aqu hay una explicacin general de cmo se relacionan:

1. **Descenso del Gradiente:**
   - El descenso del gradiente es un algoritmo de optimizacin utilizado para minimizar una funcin de costo. En el contexto de las redes neuronales, esta funcin de costo mide qu tan bien la red est realizando la tarea deseada en comparacin con los resultados reales.
   - El objetivo es ajustar los pesos y sesgos de la red para minimizar la funcin de costo. Esto se logra iterativamente actualizando los parmetros en la direccin opuesta al gradiente de la funcin de costo con respecto a esos parmetros.

2. **Retropropagacin (Backpropagation):**
   - La retropropagacin es un algoritmo especfico utilizado para calcular el gradiente de la funcin de costo con respecto a los pesos de la red. Este clculo se realiza eficientemente utilizando la regla de la cadena.
   - La retropropagacin se llama as porque el gradiente se calcula retrocediendo a travs de la red neuronal, de la capa de salida hacia la capa de entrada. Durante este proceso, se calculan los gradientes parciales para cada peso y sesgo.

3. **Conexin Terica:**
   - En el entrenamiento de una red neuronal, se utiliza el descenso del gradiente para actualizar los pesos y sesgos de la red con el objetivo de minimizar la funcin de costo.
   - La retropropagacin se utiliza para calcular el gradiente de la funcin de costo con respecto a los pesos y sesgos de la red, proporcionando as la informacin necesaria para la actualizacin de los parmetros en el descenso del gradiente.

En resumen, la retropropagacin es un componente esencial en el entrenamiento de redes neuronales, y su papel fundamental es proporcionar los gradientes necesarios para que el descenso del gradiente pueda ajustar los parmetros de la red de manera eficiente. La retropropagacin y el descenso del gradiente trabajan juntos para mejorar el rendimiento de la red neuronal durante el entrenamiento.

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Definir una red neuronal simple
model = models.Sequential([
    layers.Dense(2, input_shape=(2,), activation='relu'),  # Capa oculta con 2 neuronas y activacin ReLU
    layers.Dense(1, activation='sigmoid')  # Capa de salida con 1 neurona y activacin sigmoide para problemas de clasificacin binaria
])

# Definir una funcin de costo (por ejemplo, para un problema de clasificacin binaria)
loss_fn = tf.keras.losses.BinaryCrossentropy()

# Datos de ejemplo
X_train = tf.constant([[0.5, 0.6]])
y_true = tf.constant([[1.0]])

# Paso de entrenamiento
with tf.GradientTape() as tape:
    # Propagacin hacia adelante
    y_pred = model(X_train)

    # Calcular la prdida
    loss = loss_fn(y_true, y_pred)

# Obtener gradientes usando la retropropagacin
gradients = tape.gradient(loss, model.trainable_variables)

# Aplicar descenso del gradiente para actualizar los pesos
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
optimizer.apply_gradients(zip(gradients, model.trainable_variables))

# Imprimir los pesos antes y despus de la actualizacin
print("Pesos antes de la actualizacin:")
print(model.get_weights())

# Imprimir la prdida
print("Prdida:", loss.numpy())

# Entrenamiento completo
for _ in range(100):
    with tf.GradientTape() as tape:
        y_pred = model(X_train)
        loss = loss_fn(y_true, y_pred)

    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

# Imprimir los pesos despus del entrenamiento
print("\nPesos despus del entrenamiento:")
print(model.get_weights())

# Imprimir la prdida despus del entrenamiento
print("Prdida despus del entrenamiento:", loss.numpy())
```

```python
import numpy as np

# Funciones de activacin y sus derivadas
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Datos de entrada y salida
X = np.array([[0.5, 0.6]])
y_true = np.array([[1]])

# Parmetros de la red neuronal (inicializados aleatoriamente)
np.random.seed(42)
input_size = 2
hidden_size = 2
output_size = 1

weights_input_hidden = np.random.rand(input_size, hidden_size)
weights_hidden_output = np.random.rand(hidden_size, output_size)

# Hiperparmetros
learning_rate = 0.01
epochs = 1

# Entrenamiento de la red neuronal
for epoch in range(epochs):
    # Propagacin hacia adelante
    hidden_input = np.dot(X, weights_input_hidden)
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, weights_hidden_output)
    y_pred = sigmoid(final_input)

    # Calcular la prdida
    loss = 0.5 * np.sum((y_pred - y_true)**2)

    # Retropropagacin
    error = y_pred - y_true
    delta_output = error * sigmoid_derivative(y_pred)

    error_hidden = delta_output.dot(weights_hidden_output.T)
    delta_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Actualizar pesos con descenso del gradiente
    weights_hidden_output -= learning_rate * hidden_output.T.dot(delta_output)
    weights_input_hidden -= learning_rate * X.T.dot(delta_hidden)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Imprimir los pesos despus del entrenamiento
print("\nPesos despus del entrenamiento:")
print("Pesos input-hidden:")
print(weights_input_hidden)
print("Pesos hidden-output:")
print(weights_hidden_output)
```

La eleccin de utilizar `x * (1 - x)` en lugar de `sigmoid(x) * (1 - sigmoid(x))` en el clculo de la derivada de la funcin de activacin sigmoide durante la retropropagacin es una convencin comn. Ambas formas son equivalentes y pueden utilizarse, pero `x * (1 - x)` es una versin algebraicamente ms eficiente de la derivada de la sigmoide.

La derivada de la funcin sigmoide es:

\[ \sigma'(x) = \sigma(x) \cdot (1 - \sigma(x)) \]

donde \(\sigma(x)\) es la funcin sigmoide. Si representamos \(\sigma(x)\) como `y`, la derivada se convierte en:

\[ \sigma'(x) = y \cdot (1 - y) \]

Cuando estamos en la capa de salida de una red neuronal y ya hemos calculado la salida de la funcin sigmoide (que llamamos `y`), podemos simplemente usar `y * (1 - y)` en lugar de llamar nuevamente a la funcin sigmoide para mejorar la eficiencia computacional.

Ambas formas son correctas y darn los mismos resultados, pero `x * (1 - x)` es una forma ms eficiente de calcular la derivada si ya hemos calculado la salida de la funcin sigmoide en la capa de salida.
